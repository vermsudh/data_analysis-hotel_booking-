## DATA ANALYSIS AND VISUALIZATIONS

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc
print(cancelled_perc)
```

```python
plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled', 'Canceled'],df['is_canceled'].value_counts(), edgecolor='k', width = 0.7)
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

ax1.legend(bbox_to_anchor=(1,1))

plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')

plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel'].copy()
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel'].copy()
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')['adr'].mean().reset_index()
city_hotel = city_hotel.groupby('reservation_status_date')['adr'].mean().reset_index()
```

```python
plt.figure(figsize=(20,8))

plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')

plt.title('Average Daily Rate (ADR) Over Time', fontsize=24)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Average Daily Rate', fontsize=14)

plt.legend(fontsize=14)
plt.grid(alpha=0.3)

plt.show()
```
